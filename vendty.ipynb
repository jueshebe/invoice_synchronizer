```python
import pandas as pd
from datetime import datetime
import math
```

```python
df = pd.read_csv("ventas2.tsv", sep='\t')
```

```python
df.head()
```

```python
fv_df = df[df['Factura'].str.startswith('FVE')]
```

```python
fv_df["Factura"] = fv_df["Factura"].apply(lambda data: data.replace("FVE", ""))
```

```python
fv_df.dtypes
```

```python
def to_date(data: str) -> str:
    data = data.split(" ")[0]
    time = datetime.strptime(data, '%Y-%m-%d')
    year = time.year
    month = time.month
    day = time.day
    return f"{day}/{month}/{year}"
    
```

```python
fv_df["Fecha"] = fv_df["Fecha"].apply(to_date)
```

```python
fv_df["valor_unidatio"] = fv_df["Total venta"].apply(lambda valor: round(float(valor.split(",")[0])/1.08, 4))
```

```python
fv_df.head()
```

```python
fv_df.to_csv("cleaned.tsv", sep="\t")
```
